```python
# Define the codon table
dna_codon_table = {
    "TTT": "F", "TTC": "F", "TTA": "L", "TTG": "L",
    "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",
    "ATT": "I", "ATC": "I", "ATA": "I", "ATG": "M",
    "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",
    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S",
    "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "TAT": "Y", "TAC": "Y", "TAA": "*", "TAG": "*",
    "CAT": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
    "AAT": "N", "AAC": "N", "AAA": "K", "AAG": "K",
    "GAT": "D", "GAC": "D", "GAA": "E", "GAG": "E",
    "TGT": "C", "TGC": "C", "TGA": "*", "TGG": "W",
    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AGT": "S", "AGC": "S", "AGA": "R", "AGG": "R",
    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G"
}


# Define translation function
def translate_dna(dna_seq):
    dna_seq = dna_seq.upper()   # converts the sequence to uppercase to ensure consistency
    protein = []                # start an empty list to store translated aminoacids
    codons = [dna_seq[i:i+3] for i in range(0, len(dna_seq), 3)]  # Splits the DNA sequence into codons (3-base chunks) by taking every 3 nucleotides as one element in a list.

    for codon in codons:   # loops through each codon
        if len(codon) < 3:  # skip incomplete codons at the end
            continue
        elif codon in ['TAA', 'TGA', 'TAG']:  # stop translation if a stop codon is found
            break
        elif codon in dna_codon_table:
            protein.append(dna_codon_table[codon]) # if codon exists in the dna_codon_table, add its aminoacid to the 'protein' list
        else:
            protein.append("X")  # if unknown codon(characters other rhan A,T,G,C...like N), append 'X' to indicate an unknown amino acid

    return "".join(protein) # join all amino acids in the list into a single protein sequence string


print("Protein sequence :", translate_dna("AGGTTGCCTTTA")) # prints the translated protein for the given DNA sequence



```

```python
# In bioinformatics: Hamming distance measures the number of differing positions
# between two DNA, RNA, or protein sequences of equal length.
# useful for comparing genetic variants or detecting mutations.

# defining a function to calculate Hamming distance between two sequences
def hamming_distance(slack_name, x_name):
    if len(slack_name) != len(x_name): # Check if both strings are the same length
        raise ValueError("String mismatch")  # # Raise an error if lengths are different
    distance = sum(ch1 != ch2 for ch1, ch2 in zip(slack_name, x_name))
    # zip: pairs up corresponding characters from both strings and compare them
    # for each position, ch1 != ch2 returns True (1) if characters differ, False (0) if same
    # adding up these True values gives the total count of differing positions
    print(f"Hamming distance = {distance}")
    # prints the result showing how many characters are different


# output with 2 strings
hamming_distance("himabindu", "raghurams")
```
